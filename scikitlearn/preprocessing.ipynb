```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
 
```

```python
X,y = load_iris(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

Scaling using Standard Scaler

```python
scaler= StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

Using RobustScaler

```python
from sklearn.preprocessing import RobustScaler

scaler= RobustScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

ENCODING

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
data=fetch_openml('car',as_frame=True).frame
```

```python
data.sample(10)
```

Ordinal Encoder

```python
columns_to_encode=['lug_boot','safety','buying']

encoder=OrdinalEncoder(categories=[['small','med','big'],
                       ['low','med','high'],
                       ['low','med','high','vhigh']
                       ]
                       )
data[columns_to_encode]=encoder.fit_transform(data[columns_to_encode])
```

```python
maint_coln=OrdinalEncoder(categories=[['low','med','high','vhigh']])
data['maint']=maint_coln.fit_transform(data[['maint']])
```

```python
data['doors']=data['doors'].replace('5more',5).astype(int)
data['persons']=data['persons'].replace('more',5).astype(int)
```

```python
data=fetch_openml('adult',as_frame=True).frame
```

One hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
coln_to_encode=OneHotEncoder(handle_unknown='ignore',sparse_output=False)

encode_value=coln_to_encode.fit_transform(data[['occupation','race','relationship']])

new_coln=coln_to_encode.get_feature_names_out(['occupation','race','relationship'])

encode_df=pd.DataFrame(encode_value,columns=new_coln)

data=pd.concat([data,encode_df],axis=1)

data=data.drop(columns=['occupation','race','relationship'])
```

```python
ohot=OneHotEncoder(handle_unknown='ignore',sparse_output=False)

encode_value=ohot.fit_transform(data[['workclass','education']])

new_coln=ohot.get_feature_names_out(['workclass','education'])

encoded_df=pd.DataFrame(encode_value,columns=new_coln)

data=pd.concat([data,encode_df],axis=1)

data=data.drop(columns=['workclass','education'])
```
